```python
#implementation using arrays
class Stack:
  def __init__(self,max):
    self.max=max
    self.top=-1
    self.data=[]
  def push(self,ele):
    self.ele=ele
    if self.top+1>self.max-1:
      print("Stack overflow")
    else:
      self.top=self.top+1
      self.data.insert(self.top,ele)
  def pop(self):
    if self.top<0:
      print("Stack underflow")
    else:
      print(self.data[self.top])
      self.top=self.top-1
      #self.data.pop()
  def display(self):
    if self.top<0:
      print("Stack Underflow")
    else:
      print("The elements of Stack are:-")
      for i in range(self.top,-1,-1):
        print(self.data[i])
  def isempty(self):
    if self.top==-1:
      return True
    else:
      return False
  def isfull(self):
    if self.top>=self.max:
      return True
    else:
      return False
  def peak(self):
    return self.data[self.top]

max=int(input("Enter size of stack:"))
stc=Stack(max)
while True:
  print("\n1.Push\n2.Pop\n3.Display\n4.Exit")
  ch=int(input("Enter your choice:"))
  
  if ch==1:
    ele=eval(input("Enter element to be inserted:"))
    stc.push(ele)
  elif ch==2:
    stc.pop()
  elif ch==3:
    stc.display()
  elif ch==4:
    break
  elif ch>4:
    print("#Enter Valid Input#")
```
