```python
#queue operations using linked list.
class Node:
    def __init__(self, data):
       self.data = data
       self.next = None
 
class Queue:
    def __init__(self):
        self.head = None
        self.last = None
 
    def enqueue(self, data):
        if self.last is None:
            self.head = Node(data)
            self.last = self.head
        else:
            self.last.next = Node(data)
            self.last = self.last.next
 
    def dequeue(self):
        if self.head is None:
            return None
        else:
            to_return = self.head.data
            self.head = self.head.next
            return to_return
    def display(self):
      temp=self.head
      while temp.next!=None:
        print(temp.data,end="<--")
        temp=temp.next
      print(temp.data)
      print()
 
a_queue = Queue()
print('1.Enqueue')
print('2.Dequeue')
print('3.Display')
print('4.Quit')
while True:
  ch=int(input("\nEnter choice: "))
  if ch==1:
    data=eval(input("Enter data:"))
    a_queue.enqueue(data)
  elif ch==2:
    dequeued = a_queue.dequeue()
    if dequeued is None:
       print('Queue is empty.')
    else:
       print('Dequeued element: ', int(dequeued))
  elif ch==3:
    a_queue.display()
  elif ch==4:
    break
  else:
    print("##Enter Valid Input##")
```

```python
# queue operations using arrays
class queuem:
  def __init__(self,max):
    self.front=0
    self.rear=0
    self.queue=[]
    self.max=max
  def enqueue(self,ele):
    if len(self.queue)==self.max:
      print("Queue is full")
    else:
      self.queue.append(ele)
      self.rear+=1
  def dequeue(self):
    if len(self.queue)==0:
      print("Queue is empty")
    else:
      x=self.queue.pop(0)
      self.front+=1
      print("DELETED ELEMENT:",x)
  def isempty(self):
    if len(self.queue)==0:
      return True
    else:
      return False
  def isfull(self):
    if len(self.queue)==self.max:
      return True
    else:    
      return False
  def display(self):
    if len(self.queue)==0:
      print("No elements in Queue")
    else:
      for i in self.queue:
        print(i,end=" <--")

max=int(input("Enter size of queue:"))
print("1.ENQUEUE\n2.DEQUEUE\n3.DISPLAY\n4.EXIT")
q1=queuem(max)
while True:
  ch=int(input("\nEnter choice:"))
  if ch==1:
    ele=eval(input("Enter element:"))
    q1.enqueue(ele)
  elif ch==2:
    q1.dequeue()
  elif ch==3:
    q1.display()
  elif ch==4:
    break
  else:
    print("#wrong input#")
```
