```python
#circular single linked list
class Node:
    def __init__(self,data=None):
        self.data=data
        self.next=None
class CSLL:
    def __init__(self):
        self.head=None
    def InsertBegin(self,newdata):
        newnode=Node(newdata)
        if self.head is None:
            self.head=newnode
            newnode.next=self.head
        else:
            temp=self.head
            while temp.next!=self.head:
                temp=temp.next
            temp.next=newnode
            newnode.next=self.head
            self.head=newnode
    def InsertEnd(self,newdata):
        newnode=Node(newdata)
        if self.head is None:
            self.head=newnode
            newnode.next=self.head
        else:
            temp=self.head
            while temp.next!=self.head:
                temp=temp.next
            temp.next=newnode
            newnode.next=self.head
    def DeleteBegin(self):
        if self.head==None:
            print("No elements")
        else:
            temp=self.head
            while temp.next!=self.head:
                temp=temp.next
            temp.next=self.head.next
            self.head=temp.next
    def DeleteEnd(self):
        if self.head==None:
            print("No elements")
        else:
          prev=self.head
          temp=prev.next
          while temp.next!=self.head:
            prev=prev.next
            temp=temp.next
          prev.next=self.head
          #self.head=prev.next
    def display(self):
        temp=self.head
        while temp.next!=self.head:
            print(temp.data,end="->")
            temp=temp.next
        print(temp.data,end="->")
        print(self.head.data)

list1=CSLL()
print("""1.Insertion at begin
2.Insertion at end
3.Deletion at begin
4.Deletion at end
5.Display
6.Exit""")
while True:
  ch=int(input("\nEnter Choice: "))
  if ch==1:
    data=eval(input("Enter data:"))
    list1.InsertBegin(data)
  elif ch==2:
    data=eval(input("Enter data: "))
    list1.InsertEnd(data)
  elif ch==3:
    list1.DeleteBegin()
  elif ch==4:
    list1.DeleteEnd()
  elif ch==5:
    list1.display()
  elif ch==6:
    break
  else:
    print("## Wrong Input ##")
```
